### Fill NaN values in Age column with mean value of this column.

```python
srednia = tabela["Age"].mean()
tabela["Age"] = tabela["Age"].fillna(srednia)
tabela["Age"]
```

### Fill NaN values in Embarked with most common location.

```python
tabela["Embarked"] = tabela["Embarked"].fillna(tabela["Embarked"].value_counts().idxmax())
```

```python
tabela[tabela["Age"] < 20]
```

```python
tabela[(tabela["Age"] > 20) & (tabela["Age"] < 40)] # `&` to AND, a `|` to OR
```

### Display passengers with Age above 50

```python
tabela[tabela["Age"] > 50][["Name", "Age"]]
```

### Display passengers that Embarked at location S and are female

```python
tabela[(tabela["Embarked"] == 'S') & (tabela["Sex"] == "female")]
```

### Display passengers that paid for ticket more than 50 and are not in first class

```python
tabela[(tabela["Fare"] > 50) & (tabela["Pclass"] != 1)]
```
